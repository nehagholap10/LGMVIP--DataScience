```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
from google.colab.patches import *
```

```python
image = cv2.imread("/content/images.jpg")
#plt.figure(figsize=(3, 9.195), dpi=50)
cv2_imshow(image)
```

```python
# Converting the image into RGB format

img_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
cv2_imshow(img_rgb)
```

```python
# Converting the image into greyscale

grey_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
grey_image
```

```python
cv2_imshow(grey_image)
```

```python
# Converting the image into Inverted Image

inverted_img=255-grey_image
cv2_imshow(inverted_img)
```

```python
# Converting the image into Gaussian Blur
blurred=cv2.GaussianBlur(inverted_img,(21,21),0)
cv2_imshow(blurred)
```

```python
# Converting the image into Inverted Blurred

inverted_blurred=255-blurred
cv2_imshow(inverted_blurred)
```

```python
# Converting the image into Sketch

pencil_sketch=cv2.divide(grey_image,inverted_blurred,scale=256.0)
cv2_imshow(pencil_sketch)
```
